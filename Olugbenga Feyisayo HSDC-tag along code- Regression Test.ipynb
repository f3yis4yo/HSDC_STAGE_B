# QUESTION 13


```python
import pandas as pd
```

```python
energy_df = pd.read_csv('energydata_complete.csv')
energy_df
```

```python
simple_energy_df = energy_df.drop(columns=['date', 'lights'])
simple_energy_df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
normalised_df=pd.DataFrame(scaler.fit_transform(simple_energy_df), columns=simple_energy_df.columns)
```

```python
features_df = normalised_df.drop(columns=[ 'Appliances'])
```

```python
App_target = normalised_df['Appliances']
```

```python
App_target = pd.DataFrame(App_target)
```

```python
App_target.head()
```

```python
features_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features_df, App_target, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
predicted_values = linear_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

```python
def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(2)
    return weights_df
```
